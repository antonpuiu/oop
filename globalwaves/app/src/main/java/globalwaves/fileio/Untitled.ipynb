```python
import os
```

```python
input_dir = "input"
output_dir = "output"
```

```python
def get_files(directory):
    all_files = []

    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            all_files.append(file_path)

    return all_files
```

```python
input_files = get_files(input_dir + "/" + "command")
output_files = []
separator = 4 * " "

# pentru fiecare CommandInput
for input_file_name in input_files:
    # Creaza path asociat pentru output
    output_file = input_file_name.replace("Input", "Output").replace("input", "output")

    # Daca exista deja fisier asociat, trecem peste
    if os.path.exists(output_file):
        continue

    print(output_file)
    output_file_folders = output_file.split('/')
    class_name = output_file_folders[-1][0:output_file_folders[-1].index(".")]

    package_line = f"package globalwaves.fileio.{'.'.join(output_file_folders[:-1])};"

    arrays_import = "import java.util.Arrays;"
    list_import = "import java.util.List;"

    command_input_import = "import globalwaves.fileio.input.command.CommandInput;"
    command_output_import = "import globalwaves.fileio.output.command.CommandOutput;"
    command_result_import = "import globalwaves.fileio.output.command.CommandResult;"

    class_declaration_line = f"public class {class_name} extends CommandOutput " + "{"
    constructor_line       = separator + f"public {class_name}(CommandInput inputCommand) " + "{"
    constructor_content    = separator + separator + "super(inputCommand);"
    constructor_close      = separator + "}"

    override_adnotation = separator + "@Override"
    method_declaration = separator + "public List<CommandResult> getPossibleResults() {"
    method_content = separator + separator + f"return Arrays.asList({class_name}.Result.values());"
    method_close = separator + "}"

    enum_declaration        = separator + "public enum Result implements CommandResult {"
    enum_value              = separator + separator + "DEFAULT {"
    enum_value_override     = separator + separator + override_adnotation
    enum_value_decl         = separator + separator + separator + "public String getResult(String arg) {"
    enum_value_content      = separator + separator + separator + separator + "return null;"
    enum_value_func_close   = separator + separator + separator + "}"
    enum_value_close        = separator + separator + "}"
    enum_close              = separator + "}"
    

    class_close = "}"

    lines = [
        package_line,
        "",
        arrays_import,
        list_import,
        "",
        command_input_import,
        command_output_import,
        command_result_import,
        "",
        class_declaration_line,
        constructor_line,
        constructor_content,
        constructor_close,
        "",
        override_adnotation,
        method_declaration,
        method_content,
        method_close,
        "",
        enum_declaration,
        enum_value,
        enum_value_override,
        enum_value_decl,
        enum_value_content,
        enum_value_func_close,
        enum_value_close,
        enum_close,
        class_close
    ]

    directories, filename = os.path.split(output_file)
    if directories:
        os.makedirs(directories, exist_ok=True)

    with open(output_file, "w") as write_file:
        write_file.write("\n".join(lines))

    print("\n".join(lines))
    # break
```
